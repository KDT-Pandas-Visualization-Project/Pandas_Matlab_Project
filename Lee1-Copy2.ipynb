# 원작(소설 / 만화)이 있는 경우와 있는 경우 평점의 차이가 있을까?
- 사용 데이터 : anime 데이터
- 사용 column : 'anime_id', 'Name', 'Score', 'Source', 'Rating', 'Rank', 'Popularity',  'Favorites', 'Scored By
 .

(uint)           Rank:        인기도나 기타 기준에 의한 애니메이션의 랭크.  
(int)            popularity:  애니메이션의 인기 순위.  
(int)            Favorites:   애니메이션이 사용자가 즐겨찾기로 표시한 횟수

```python
sourceDF = anime[['anime_id', 'Name', 'Score', 'Source', 'Rank', 'Popularity', 'Favorites', 'Scored By']].set_index("anime_id")
```

```python
# Scored By가 0이 아닌 데이터만 사용할 것 -> 평균치가 너무 낮아짐 
sourceDF = sourceDF[sourceDF["Scored By"] != 0]
sourceDF
```

```python
sourceDF["Source"].unique()
```

```python
groupDF = sourceDF.groupby(["Source"]).mean(numeric_only=True)
groupDF
```

```python
groupDF.info()
```

```python
groupDF2=groupDF.reset_index()
```

```python
retDF=groupDF2.Source.astype('object').str.split(expand=True)
retDF
```

```python
retDF3=pd.concat([groupDF2, retDF], axis='columns').set_index([1, "Source"])
```

```python
retDF3 = retDF3.sort_index()
retDF3.drop(0, axis=True, inplace=True) # 불필요한 col 삭제 
```

```python
retDF3
```

```python
# manga col과 novel col을 나눈다
manga_col = []
novel_col = []
for idx in groupDF.index:
    # print(type(idx)) -> str형
    if "manga" in idx.lower():
        manga_col.append(idx)
    if "novel" in idx.lower():
        novel_col.append(idx)
print(manga_col)
print(novel_col)
```

```python
# 소설 
novelDF =groupDF.loc[novel_col]
novelDF
```

```python
novelSR = pd.DataFrame(novelDF.mean()).T
novelSR.index = ["novel"]
novelSR
```

```python
# 망가 
mangaDF = groupDF.loc[manga_col]
mangaDF
```

```python
mangaSR = pd.DataFrame(mangaDF.mean()).T
mangaSR.index = ["manga"]
mangaSR
```

```python
oriSR = pd.DataFrame(groupDF.loc["Original"]).T
oriSR.index.name = "Source"
oriSR
```

```python
rankDF = pd.concat([novelSR, mangaSR, oriSR])
rankDF
```

```python
rankDF = rankDF.drop(["Popularity", "Favorites"], axis=1)
```

```python
rankDF.corr()
```

```python
plt.matshow(rankDF.corr())
plt.show()
```

```python
rankDF.loc["Original"]
```

```python
index = [0, 0.5, 1]
label = ["novel", "manga", "original"]
p1 = plt.bar(0, rankDF.loc["novel", "Score"], color='red', alpha=0.5, width=0.4)
p2 = plt.bar(0.5,rankDF.loc["manga", "Score"], color='blue', alpha=0.5, width=0.4)
p3 = plt.bar(1,rankDF.loc["Original", "Score"], color='yellow', alpha=0.5, width=0.4)
plt.plot(index, rankDF["Score"], 'ko-', label='line 1', linewidth=2)
y = np.arange(0, 8, 0.5)
ylabel = [0, "", 1, "", 2, "", 3, "", 4, "", 5, "", 6, "", 7, ""]
plt.xticks(index,label) 

plt.yticks(y,ylabel)
plt.grid(True, lw=1, ls='--', c='.75')
plt.legend((p1[0], p2[0], p3[0]), label, fontsize=15)
plt.title(f"원작 별 평점 max = 10")
plt.xlabel("원작 종류")
plt.ylabel("평점 평균")

a = -0.05
plt.annotate(round(rankDF.loc["novel", "Score"], 2),(0+a, 7))
plt.annotate(round(rankDF.loc["manga", "Score"], 2),(0.5+a, 7))
plt.annotate(round(rankDF.loc["Original", "Score"], 2),(1+a, 6.5))
```
